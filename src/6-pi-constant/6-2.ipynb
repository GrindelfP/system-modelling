```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Time:
    def __init__(self, time_as_minutes):
        self.time_as_minutes = time_as_minutes
        self.common_time_format = self.to_common()
    
    
    def to_common(self) -> str:
        hours, minutes = divmod(self.time_as_minutes, 60)    
        return "{:02d}:{:02d}".format(hours, minutes)
    
    def show_time(self) -> None:
        print(self.common_time_format)
```

```python
time_limits: list[int] = [10, 15, 10, 5, 3, 40, 5, 7]
time_errors: list[int] = [2, 3, 5, 3, 2, 5, 3, 3]
class_time: Time = Time(770)
iterations: int = 10000000
```

```python
def get_total_time_minutes() -> int:
    total_time: int = 0
    for i in range(0, len(time_limits)):
        time_limit = time_limits[i]
        time_error = time_errors[i]
        total_time += np.random.randint(time_limit - time_error, time_limit + time_error + 1)
    
    return total_time
```

```python
times: list[int] = []
for i in range(iterations):
    times.append(get_total_time_minutes())
```

```python
k95: int = int(iterations * 0.95)
times.sort()
upper_time_bound_minutes: int = times[k95]
```

```python
wake_up_time: Time = Time(class_time.time_as_minutes - upper_time_bound_minutes)
wake_up_time.show_time()
```

```python
def plot_histogram(numbers):
    plt.hist(numbers, bins='auto', alpha=0.7, color='green', edgecolor='black')
    plt.title('Гистограмма частоты различных значений времени действий студента')
    plt.xlabel('Время')
    plt.ylabel('Частота')
    plt.show()
```

```python
plot_histogram(times)
```
